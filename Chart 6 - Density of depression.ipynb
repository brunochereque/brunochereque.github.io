```python
#Import the package to your work environment
import fingertips_py as ftp
import pandas as pd
import altair as alt
```

```python
# Define the Indicator ID: Depression incidence
depression = 848 

# Fetch the data for the specified indicator at all available geographies
depression_data = ftp.get_data_for_indicator_at_all_available_geographies(depression)

# Display the first few rows of the data to understand the structure
depression_data.head(2)
```

```python
# Inspect columns
depression_data.columns
depression_data["Area Type"].unique()
depression_data_counties = depression_data[depression_data["Area Type"]=="Districts & UAs (from Apr 2023)"]

depression_data_counties["Area Type"].unique()

depression_data_counties["Time period"].unique()

depres = depression_data_counties[depression_data_counties["Time period"]=="2022/23"]

depres["Time period"].unique()
```

```python
depres.columns
depres = depres[['Value']]
```

```python
import altair as alt

# Histogram
hist = alt.Chart(depres).mark_bar(opacity=0.5, color="steelblue").encode(
    alt.X('Value:Q', bin=alt.Bin(maxbins=30), title="Depression Prevalence (%)"),
    alt.Y('count()', title="Count"),
)

# KDE (adjusted for dual axis)
kde = alt.Chart(depres).transform_density(
    'Value',
    as_=['Value', 'Density']
).mark_line(color="red", strokeWidth=3).encode(
    alt.X('Value:Q', title="Depression Prevalence (%)"),
    alt.Y('Density:Q', title="Density")
)

# Combine the histogram and KDE, with dual axes
chart = alt.layer(hist, kde).resolve_scale(
    y='independent'  # Allow the histogram and KDE to use independent y-axes
).properties(
    title={
        "text": "Distribution of Depression Prevalence (2023)",
        "subtitle": "Based on District-Level Depression Prevalence | Histogram with KDE Overlay (Dual Axis)",
        "fontSize": 20,
        "subtitleFontSize": 14,
        "anchor": "start"
    },
    width=800,
    height=400
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_title(
    fontSize=18,
    subtitleFontSize=14,
    anchor='start'
)

chart


#chart.save('Chart_6_histogram.json')
```
